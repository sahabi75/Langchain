""String Prompt""

```python
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate.from_template("What year is it?")
```

```python
from langchain_google_genai import GoogleGenerativeAI
llm = GoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.9)
response = llm.invoke(prompt.format())
print(response)
```

##Chatpromt templet


```python
from langchain_core.prompts import ChatPromptTemplate
chat_prompt = ChatPromptTemplate([
    ("system", "You are a helpful assistant."),
    ("user", "What is the capital of {country}?"),
])
prompt = chat_prompt.invoke({"country": "France"})
response = llm.invoke(prompt)
print(response)
```

##MessagePlaceHolder##


```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    MessagesPlaceholder("msg")])
prompt = prompt.invoke({"msg": [HumanMessage(content="What is the capital of France?")]})
response = llm.invoke(prompt)
print(response)
```
