##Sequence Chain

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate, MessagesPlaceholder
from json import loads
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
load_dotenv()

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.9)
parser = StrOutputParser()
template1 = PromptTemplate.from_template("""
    give me a short summary on topic: {topic}
""")
template2 = PromptTemplate.from_template("""
Based on the following summary, give me 4 key points on it {summary}
""")

seq_chain = template1 | llm | parser | template2 | llm | parser
response = seq_chain.invoke({"topic": "Artificial Intelligence"})
print(response)
```
