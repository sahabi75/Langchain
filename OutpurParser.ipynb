```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate
parser = JsonOutputParser()
template = PromptTemplate.from_template("""
Extract the following information from the text and return it as a JSON
- Name
- Price
- Description
- Availability
- Reviews
-Realease Date
-features(list)
text: {text}
format_instructions: {format_instructions}
""",
partial_variables={"format_instructions": parser.get_format_instructions()}

)

chain = template | llm | parser
response = chain.invoke({"text": "The iPhone 14 Pro Max is the latest flagship smartphone from Apple, released in September 2022. It's price is 140000tk. It features a stunning 6.7-inch Super Retina XDR display, powered by the A16 Bionic chip for lightning-fast performance. The phone boasts a triple-camera system with advanced computational photography capabilities, allowing users to capture stunning photos and videos. With its sleek design, long battery life, and 5G connectivity, the iPhone 14 Pro Max is a top choice for tech enthusiasts and Apple fans alike."})
print(response)
```
